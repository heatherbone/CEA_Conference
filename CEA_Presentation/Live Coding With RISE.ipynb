### Motivation


- Economics students increasingly need programming skills. 
    - They need to learn how to turn an economic question into a sequence of coding steps.

- Pre-written (static) code can make programming look like memorization.

- **Live coding:** Models the process by building code live and narrating the decisions behind each step.

- **This presentation:** a technical setup for combining slides, live code, output, and exercises in one interactive file.

### Standard Implementation and Challenges

- Commonly taught using Jupyter notebooks, which combine explanations and code in one file. 
    - Notebooks are not presentation friendly because they are designed for scrolling, not presenting.

- Switching between slides and a coding environment improves structure, but separates explanation from execution

### A Better Approach: Live Coding with RISE

- RISE is an extension that turns a Jupyter notebook into a slide deck

- One notebook, two uses:
    - Students recieve a guided notebook with blanks to fill in
    - Instructors present the same notebook and code the solutions live

### Demo: Introducing Loops to Economics Majors 

Suppose I want to plot a demand curve defined by:

$P = 100-4Q$


For this, I need a list of quantities and corresponding prices. 

**Approach 1:** Manual Approach

```python
#not live coded
quantities = []
prices = []

#add quantity of 0 and corresponding price to a list
quantity = 0
price = 100 - 4 * quantity
quantities.append(quantity)
prices.append(price)

quantity = 2
price = 100 - 4 * quantity
quantities.append(quantity)
prices.append(price)

quantity = 5
price = 100 - 4 * quantity
quantities.append(quantity)
prices.append(price)

print(quantities)
print(prices)
```

- **Downside:** Error prone and lengthy code

**Approach 2:** Use a *loop* to calculate the price for each quantity

```python
quantities = [0, 2, 5]
prices = []

#LIVE CODED
for quant in quantities:
    price = 100 - 4 * quant
    prices.append(price)
    
print(quantities)
print(prices)
```

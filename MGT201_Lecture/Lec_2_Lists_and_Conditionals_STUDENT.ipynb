Indexing and slicing works for lists similar to strings.

````python
mylist = [100, 350, 720, 2500, 7800, 9000] 
mylist[2]
#UNCOMMENT THE LISTS BELOW TO REVEAL THEM
#mylist[1:]
#mylist[:5]
````

Like strings, we can use **[start:end:increment]** to access a sublist of a list. 

If we omit one of the values, then the default rule will be imposed. You can play with it to see if the default rule is consistent with your expectation.


````python
mylist[::2]

#UNCOMMENT THE LISTS BELOW TO REVEAL THEM

#mylist[3::2]

#mylist[:2]
````

We can also use a **Negative index** in `[]`. 

Let's learn the meaning by looking at the following examples: 

````python
mylist = [100, 350, 720, 2500, 7800, 9000] 
mylist[-1]
````

As we can see, it outputs the last element. Importantly, negative indexing does not start at zero. 


Let's try the following:

````python
mylist[-1:3]
````

It returns an empty list. **Why is that?**
- When you do this, you're asking Python: “Start at the last element and move forward until you reach index 3.”

**But there's a problem:**
- Index -1 is after index 3 in the direction of slicing (since slicing goes left-to-right).
- So Python says: “There's nothing in that forward range from -1 to 3.”

In the code below, we set the step size to be negative.



````python
mylist[0:3:-1]
````

It's an empty list. **Why?**
- You're telling Python: “Start at index 0, and move backward, stopping before you hit index 3.” 

**That’s a contradiction** — you’re starting at the beginning of the list, but trying to move backward, toward the left… and there’s nothing before index 0. 

To actually slice from right to left, the later index must be the starting index and the earlier index the ending index, and we must use a negative step size. 

````python
mylist = [100, 350, 720, 2500, 7800, 9000] 
mylist[3:0:-1]
````

For nested lists, the indexing rule follows similarly with double `[]`.
- The first set of brackets selects the outer list, and the second set selects an item from the inner list at that position.

### Change Values in a List 

Recall that the object types covered in lecture 1 (numbers, booleans, and strings) were **immutable objects**. 
- This means that we cannot change objects directly
- We must create new objects and then change what the objects refer to

````python
#EXAMPLE FOR REVIEW HERE
price = 10
id(price)
price = 15
id(price)
````

Unlike the objects we covered last lecture, lists are **mutable**
- We can update their values in place without reassinging the list to another variable

The underlying object in the memory remains the same. 

````python
mylist = ['a', 'b', 'c']
#LIVE CODED - #replace element b with the number 30
````

We can confirm this is what is happening by printing out the ID of `mylist`. It's pointing to the same object after updating. 

````python
mylist = ['a', 'b', 'c']
id(mylist)
mylist[1] = 30
id(mylist)
mylist
````

We can even change slices of lists. To do so, we select part of the list and assign new values to it. 
- Again, the syntax below updates the object in place. 

````python
mylist = ['a', 'b', 'c']
id(mylist)
mylist[0:2] = [30, 5]
print(mylist)
id(mylist)
````

To **remove** an element from the list, we can use `del`.

````python
mylist = [1,2,3]
del mylist[1] # delete the second element
mylist
````

#### Reassignment vs. Mutation

We *can* still use variable reassignment for lists - remember, though, that behind the scenes, this is creating a new object (and thus the id changes):

````python
mylist = ['a', 'b', 'c']
id(mylist)
mylist = ['a', 30, 'c']
id(mylist)
mylist
````

Compare this to mutation (updating the object itself):

````python
mylist = ['a', 'b', 'c']
id(mylist)
mylist[1] = 30
id(mylist)
mylist
````

Notice that the id remained the same. 

Understanding the difference can help us avoid some unexpected behaviour. One case in which this really matters is when we assign the value stored in one variable to another one. 

````python
a = ['a', 'b', 'c']
b = a # b refers to the same list
id(a)
id(b)
````

Notice that `a` and `b` have the same id. This is like having two names refer to the same object. It's like if I had two names, my name and a nickname. 

One reason why it's important to understand mutation vs. variable reassingment is that it can help us understand what happens to the value of `b` when we make changes that affect `a`. 

If we update the object assigned to `a` in place (through mutation) we are changing the object itself. As a result, when we use `b`, its associated value changes as well:

````python
a = ['a', 'b', 'c']
b = a # b refers to the same list
id(a)
id(b)
a[1] = 30
id(a)
id(b)
print(a)
print(b)
````

Note that:
- The values assigned to both `a` and `b` change. 
- The ids do not, as we are changing the object itself. 

Let's compare this to what happens when we use variable reassignment to change what `a` refers to:

````python
a = ['a', 'b', 'c']
b = a # b refers to the same list
id(a)
id(b)
a = ['a', 30, 'c']
id(a)
id(b)
print(a)
print(b)
````

What happens is different. Here, we are creating a new object and making `a` refer to it. `b` still refers to the value it got attached to when we initialized it. 
- This is why the id of `a` stayed the same and the value of `b` did not

There are other important implications of mutability versus reassignment. We'll explore these as they come up throughout the course.

## Conditional Statements

In many cases, we want Python to run different code if some conditions are satisfied.

We can control which code is run using conditional statements, which use the keywords `if`, `elif`, and `else` to choose between different blocks of code.

An `if` statement is the simplest kind of conditional: it lets Python either run a block of code (if a condition is satisfied) or skip it (if it is not).

`if` statements use the following syntax:
    
```python
if test1: #test1 is some condition (True or False)
    statement1 #code run if condition is satisfied
statement2 #code run regardless of whether condition is satisfied
```

Let's look at an example below:

````python
x = 6
#LIVE CODED - TEST IF x > 5
#if true print 'x is greater than 5'
#print 'Program ended' regardless
````

Let's dissect the above code:

- `x=6` represents the code before the *if statement*
- The if statement starts with `if`.
- `x>5` is a boolean value (it is `True` in this case). It is followed by `:`.
- `print('x is greater than 5')` is executed because `test1` is `True`.
- `print('Program ended')` is always executed regardless of the value of `test1` because it is after the if statement.

`if` is often paired with `else`. It provides more control over the flow of the program.

```python
if test1: #test1 is some condition (True or False)
    statement1  #code runs if test1 is True
else: #this line only runs if test1 isn't True
    statement2 #this line only runs if test1 isn't True
statement3 #this line runs regardless of the truth of test1
```

````python
x = 6
if x>5:
    print('x is greater than 5')
else:
    print('x is not greater than 5')
print('Program ended')
````

If we have more than a binary test (i.e, there are more than two possible outcomes), what can we do? 
- We need `elif` (short for else if). 
- It is placed between `if` and `else`. 

General syntax

```python
if test1:
    statement1
elif test2: #we can add multiple elif conditions
    statement2
else:
    statement3
statement4
```

As an example, see the following program, in which we print the number of days in a month in a given year. 

````python
current_month = 4 #line 1
if current_month in [4, 6, 9, 11]: #line 2
    print("This month has 30 days") #line 3
elif current_month == 2: #line 4
    print("This month has 28 (or 29) days") #line 5
else: #line 6
    print("This month has 31 days") #line 7
print('Done') #line 8 
````

The running order of the lines: 1->2->3->8
- Line 1 is run first to initialize `current_month`.
- Line 2 is run next to check if `current_month` is in the provided list.
- Since `current_month` points to the value of `4`, the statement is true and Line 3 is run.

### Nested Conditional Statements

We can nest if statements. Consider the following code:

````python
#Nested if - Check if an food item is a brunch item

breakfast = ['eggs', 'bacon','sandwich'] #line 1
lunch = ['burger', 'sandwich', 'pizza'] #line 2
food = 'eggs' #line 3
if food in breakfast: #line 4
    if food in lunch: #line 5
        print(f'{food} is a brunch item') #line 6
    else: #line 7
        print(f'{food} is a breakfast item') #line 8 
else: #line 9
    if food in lunch: #line 10 
        print(f'{food} is a lunch item') #line 11
    else: #line 12
        print(f'{food} is not on the menu') #line 13
````

After initializing the variables, we first check if `food` is in the list of breakfast foods (`breakfast`) on line 4:
- If it is, we enter into the block of code that runs when the first `if` condition is satisfied (lines 4-8)
- First, we test if the food is a brunch food by checking if it is in `lunch` (the list of lunch foods) (line 5)
  - If it is, we run the statement on line 6 (because it's both a breakfast and a lunch food, it's a brunch food)
    - If it isn't, we run line 7 and line 8 (it's a breakfast food that isn't a lunch food)


If the food is not a breakfast food, then the statement on liner 4 is `False`, so we skip to line 9 and enter the block of code for the outer `else` condition. 
- We check if the food is in `lunch` next (line 10).
  - If it is, we print a statement indicating that it is a lunch item (line 11)
  - If it is not, we run line 12 and line 13. This is run when the food is neither a lunch or breakfast item, indicating it is not on the menu.

We could also rewrite the code above using boolean statements and avoid nesting our conditional statements:

````python
breakfast = ['eggs', 'bacon', 'sandwich']
lunch = ['burger', 'sandwich', 'pizza']

food = 'eggs'

if food in breakfast and food in lunch:
    print(f'{food} is a brunch item')
elif food in breakfast: #only run if initial condition not satisfied
    print(f'{food} is a breakfast item')
elif food in lunch: #only run if initial condition not satisfied
    print(f'{food} is a lunch item')
else: #only run if initial condition not satisfied
    print(f'{food} is not on the menu')
````
